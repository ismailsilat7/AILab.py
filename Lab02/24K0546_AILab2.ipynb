```python
# Task 1: Student Profile System (Basics of Class & Object)
class Student:
    def __init__(self, name, roll_no, program):
        self.name = name
        self.roll_no = roll_no
        self.program = program
    def display_info(self):
        print(f"Name: {self.name}, Roll Number: {self.roll_no}, Program: {self.program}")

s1 = Student("Ismail", "24K-0546", "BSCS")
s2 = Student("Ali", "24K-0802", "BSCS")

s1.display_info()
s2.display_info()
```

```python
# Task 2: Smart Fan Controller (Constructor & Methods)
class SmartFan:
    def __init__(self, room_name, speed):
        self.room_name = room_name
        self.speed = speed
        if self.speed > 100:
            self.speed = 100
        elif self.speed < 0:
            self.speed = 0

    def increase_speed(self):
        
        self.speed += 5
        if self.speed > 100:
            self.speed = 100

    def decrease_speed(self):
        self.speed -= 5
        if self.speed < 0:
            self.speed = 0

    def display_status(self):
        print(f"Name: {self.room_name}, Speed: {self.speed}")

s1 = SmartFan("Lab 3", 50)
s1.display_status()
s1.increase_speed()

s2 = SmartFan("R-109", 50)
s2.display_status()
s2.decrease_speed()

s1.display_status()
s2.display_status()
```

```python
# Task 3: Library Book System (Class vs Instance Attributes)
class Book:
    library_name = "EE Library"
    def __init__(self, title, author):
        self.title = title
        self.author = author
    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author} - (Library {Book.library_name})")

b1 = Book("Meow", "Cat")
b2 = Book("Bhau", "Dog")
b1.display_info()
b2.display_info()

print("Changing library name for Book")
Book.library_name = "AB2 Library"

b1.display_info()
b2.display_info()
```

```python
# Task 4: Bank Account Manager (Methods & Attributes)
class BankAccount:
    def __init__(self, account_holder, account_no, balance):
        self.account_holder = account_holder
        self.account_no = account_no
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    def display_account(self):
        print("Account Holder:", self.account_holder)
        print("Account Number:", self.account_no)
        print("Balance:", self.balance)

account1 = BankAccount("Ali", 101, 50000)
account2 = BankAccount("Ismail", 102, 3000)

account1.display_account()
account1.deposit(1500)
account1.withdraw(50000)
account1.display_account()
print("--------------------------")
account2.display_account()
account2.deposit(1000000)
account2.withdraw(500)
account2.display_account()
```

```python
# Task 5: Smart Home Light System (OOP Automation)
class SmartLight:
    def __init__(self, room):
        self.room = room
        self.status = False

    def get_status(self):
        return "On" if self.status else "Off"

    def turn_on(self):
        self.status = True

    def turn_off(self):
        self.status = False

    def show_status(self):
        print(f"Room: {self.room}, Light Status: {self.get_status()}")

living_room = SmartLight("Living Room")
bedroom = SmartLight("Bedroom")

living_room.turn_on()
bedroom.turn_off()

living_room.show_status()
bedroom.show_status()

print("------------------------------------")

bedroom.turn_on()
living_room.turn_off()

living_room.show_status()
bedroom.show_status()
```

```python
# Task 6: University Staff System (Inheritance)
class Staff:
    def __init__(self, name, staff_id, department):
        self.name = name 
        self.staff_id = staff_id
        self.department = department
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Staff ID: {self.staff_id}")
        print(f"Department: {self.department}")

class Teacher(Staff):
    def __init__(self, name, staff_id, department, courses, salary):
        super().__init__(name, staff_id, department)
        self.courses = courses
        self.salary = salary

    def display_info(self):
        super().display_info()
        print(f"Courses: {self.courses}")
        print(f"Salary: {self.salary}")

class AdminStaff(Staff):
    def __init__(self, name, staff_id, department, role, working_hours):
        super().__init__(name, staff_id, department)
        self.role = role
        self.working_hours = working_hours

    def display_info(self):
        super().display_info()
        print(f"Role: {self.role}")
        print(f"Working Hours: {self.working_hours}")

class ResearchAssistant(Staff):
    def __init__(self, name, staff_id, department, research_topic, stipend):
        super().__init__(name, staff_id, department)
        self.research_topic = research_topic
        self.stipend = stipend

    def display_info(self):
        super().display_info()
        print(f"Research Topic: {self.research_topic}")
        print(f"Stipend: {self.stipend}")

teacher1 = Teacher(
    name="Ali",
    staff_id="T101",
    department="Computer Science",
    courses=["OOP", "DSA"],
    salary=120000
)

admin1 = AdminStaff(
    name="Silat",
    staff_id="A201",
    department="Administration",
    role="Office Manager",
    working_hours=8
)

ra1 = ResearchAssistant(
    name="Ismail",
    staff_id="R301",
    department="Physics",
    research_topic="Quantum Computing",
    stipend=50000
)

print("Teacher Info:")
teacher1.display_info()

print("\nAdmin Staff Info:")
admin1.display_info()

print("\nResearch Assistant Info:")
ra1.display_info()
```

```python
# Task 7: Vehicle System (Method Overriding – Polymorphism)
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

class Car(Vehicle):
    def start(self):
        print("Car is starting...")

class Bike(Vehicle):
    def start(self):
        print("Bike is starting...")
        
class Bus(Vehicle):
    def start(self):
        print("Bus is starting...")

v = Vehicle()
c = Car()
b = Bike()
bus = Bus()

v.start()
c.start()
b.start()
bus.start()
```

```python
# Task 8: Calculator Utility (Method Overloading Simulation)
class Calculator:
    def multiply(self, first=1, *args):
        result = first
        for num in args:
            result *= num
        return result
calc = Calculator()

print(calc.multiply(2, 3))
print(calc.multiply(2, 3, 4))
print(calc.multiply(2, 3, 4, 5))
print(calc.multiply())  
```

```python
# Task 9: Employee Payroll System (Protected Members)
class Employee:
    def __init__(self, name, salary):
        self._name = name
        self._salary = salary


class Manager(Employee):
    def display_info(self):
        print(f"Name: {self._name}")
        print(f"Salary: {self._salary}")

mgr = Manager("Ali", 150000)
mgr.display_info()
```
