```python
# task 6
myDict = {
    "name": "Ismail",
    "marks": 89.6,
    "grade": ""
}

if myDict["marks"] >= 90:
    myDict["grade"] = "A+"
elif myDict["marks"] >= 80:
    myDict["grade"] = "A"
elif myDict["marks"] >= 70:
    myDict["grade"] = "B"
elif myDict["marks"] >= 60:
    myDict["grade"] = "C"
elif myDict["marks"] >= 50:
    myDict["grade"] = "D"
else:
    myDict["grade"] = "E"

print(myDict)
```

```python
# task 7
import math
num = int(input("Enter a number: "))
isPrime = True
for i in range(2, int(math.sqrt(num) + 1)):
    if num % i == 0:
        isPrime = False
        break

print(f"{num} is Prime" if isPrime else f"{num} is not Prime")
```

```python
# task 10
def maxVal(*numbers):
    if len(numbers) == 0:
        return -1    
    maxV = numbers[0]
    for i in range(1, len(numbers)):
        if(numbers[i] > maxV):
            maxV = numbers[i]
    return maxV

print(maxVal(4,2,7,0))

def printKeyVal(**names):
    print("Names: ", names)

printKeyVal(n1 = "Ismail", n2 = "Jabbar", n3 = "Hammad")
```
