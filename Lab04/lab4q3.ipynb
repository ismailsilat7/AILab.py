```python
class EightPuzzleAgent:
    def __init__(self, start, goal):
        self.start = start
        self.goal = goal

    def goal_test(self, state):
        return state == self.goal

    def actions(self, state):
        space_index = None
        for i in range(3):
            for j in range(3):
                if state[i][j] == 0:
                    space_index = (i, j)
                    break;
            if space_index:
                break;
        moves = []
        i,j = space_index
        # move up
        if i > 0:
            moves.append(self.move(state, i, j, i-1, j))
        # move down
        if i < 2:
            moves.append(self.move(state, i, j, i+1, j))
        # move left
        if j > 0:
            moves.append(self.move(state, i, j, i, j-1))
        # move right
        if j < 2:
            moves.append(self.move(state, i, j, i, j+1))
        return moves

    def move(self, state, i, j, newI, newJ):
        new_state = [list(row) for row in state]
        new_state[i][j], new_state[newI][newJ] = new_state[newI][newJ], new_state[i][j]
        return tuple(tuple(row) for row in new_state)
    
    def solve(self):
        visited = set()
        q = []
        q.append(self.start)
        while q:
            n = q.pop(0)
            visited.add(n)
            if n == self.goal:
                print("Goal found!")
                print(n)
                return True
            else:
                for action in self.actions(n):
                    if action not in visited:
                        q.append(action)
        return False
```

```python
solver = EightPuzzleAgent(
    ( (7, 2, 4), (5, 0, 6), (8, 3, 1) ),
    ( (0, 1, 2), (3, 4, 5), (6, 7, 8) )
)

found = solver.solve()
if found:
    print("Found")
else:
    print("Not Found")
```
