```python
class Node:
    def __init__(self, state):
        self.state = state
        self.neighbors = []  # arr of tuples with format as (cost, Node) 
        self.parent = None

class Graph:
    def __init__(self):
        self.nodes = {}
        search_map = {
            'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
            'Zerind': [('Arad', 75), ('Oradea', 71)],
            'Oradea': [('Zerind', 71), ('Sibiu', 151)],
            'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
            'Timisoara': [('Arad', 118), ('Lugoj', 111)],
            'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
            'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
            'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
            'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
            'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
            'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
            'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
            'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
            'Giurgiu': [('Bucharest', 90)],
            'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
            'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
            'Eforie': [('Hirsova', 86)],
            'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
            'Iasi': [('Vaslui', 92), ('Neamt', 87)],
            'Neamt': [('Iasi', 87)]
        } # adjacency list generated via gemini - not cross checked

        for city in search_map:
            self.nodes[city] = Node(city)

        for city, neighbors in search_map.items():
            currNode = self.nodes[city]
            for neighbor, cost in neighbors:
                neighborNode = self.nodes[neighbor]
                currNode.neighbors.append((cost, neighborNode))

    def printPath(self, node):
        path = []
        currNode = node
        while currNode:
            path.append(currNode.state)
            currNode = currNode.parent
        print(f"Path to {node.state}: {' -> '.join(reversed(path))}")

    def UCS(self, start="Arad", target="Bucharest"):
        startN = self.nodes[start]
        visited = []
        parent = []
        q = [(0, startN)]
        while q:
            min_index = 0
            for i in range(len(q)):
                if q[i][0] < q[min_index][0]:
                    min_index = i
            currCost, node = q.pop(min_index)
            visited.append(node)
            if node.state == target:
                print("Cost: ", currCost)
                self.printPath(node)
                return True
            else:
                for cost, neighbor in node.neighbors:
                    if neighbor not in visited:
                        neighbor.parent = node
                        q.append((cost + currCost, neighbor))
        return False

searchGraph = Graph()
found = searchGraph.UCS("Arad", "Bucharest")
if found == False:
    print("No such path exists...")
```
