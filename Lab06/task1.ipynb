```python
import heapq

graph = {
    'S': [('A', 3), ('B', 5), ('C', 1)],      
    'A': [('D', 4), ('E', 5), ('F1', 2)],     
    'B': [('F', 6), ('G', 7), ('H1', 3)],
    'C': [('H', 1), ('I1', 2)],               
    'D': [('I', 2), ('J1', 3)],               
    'E': [('J', 2), ('K1', 3)],
    'F1': [('L1', 2)],
    'F': [('K', 1)], 'G': [('L', 1)], 'H1': [('M1', 2)],
    'H': [('M', 1), ('N1', 2)],               
    'I1': [('O1', 2)],
    'I': [], 'J': [], 'K': [], 'L': [], 'M': [],
    'J1': [], 'K1': [], 'L1': [], 'M1': [], 'N1': [], 'O1': []
}

def beam_search(start, goal, width=2, maxwidth=5):
    states = [ (0, [start]) ]
    level = 0
    while states:
        if level > 0 and level % 3 == 0 and width < maxwidth:
            width += 1
        print(f"Level: {level}, Width: {width}\nBeam Nodes: {states}")
        all_neighbors = []
        for cost, path in states:
            curr_node = path[-1]
            if curr_node == goal:
                return path, cost
            else:
                neighbors = graph[curr_node]
                for neighbor, edge_cost in neighbors:
                    new_cost = cost + edge_cost
                    new_path = path + [neighbor]
                    all_neighbors.append((new_cost, new_path))
        states = heapq.nsmallest(width, all_neighbors, key=lambda x:x[0])
        level += 1
    return None, float('inf')

start = 'S'
goal = 'M'
path, cost = beam_search(start, goal)

if path:
    print(f"Path found with cost {cost}\nPath: {path}")
else:
    print("No Path found")
    
```
